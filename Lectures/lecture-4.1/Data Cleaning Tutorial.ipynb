### Please clean up the book title and save the results to a new dataframe table

```python
def clean_title(title):
    
    if title == 'nan':
        return 'NaN'
    
    if title[0] == '[':
        title = title[1: title.find(']')]
        
    if 'by' in title:
        title = title[:title.find('by')]
    elif 'By' in title:
        title = title[:title.find('By')]
        
    if '[' in title:
        title = title[:title.find('[')]

    return title

#Get the clean title
df['Title'] = df['Title'].apply(clean_title)
df.head()
```

### In class Practice: 
1. Please use .merge or .join combine olympics data part1, part2, part3
2. Calculate the total number of accumulated medals of Summer and Winter Games

```python
# Step 1: Merge all three dataframes using .merge()
# First merge newdf (which contains part1 and part2 concatenated) with part3 on 'Country'
olympics_combined = newdf.merge(olympics_part3_df, on='Country', how='outer')
olympics_combined
```

```python
# Step 2: Calculate total accumulated medals from Summer and Winter Games
# Total Summer medals = Gold + Silver + Bronze (from Summer Olympics)
# Total Winter medals = Gold.1 + Silver.1 + Bronze.1 (from Winter Olympics)
# Total Combined medals = Sum of Summer and Winter medals

olympics_combined['Total_Summer_Medals'] = olympics_combined['Gold'] + olympics_combined['Silver'] + olympics_combined['Bronze']
olympics_combined['Total_Winter_Medals'] = olympics_combined['Gold.1'] + olympics_combined['Silver.1'] + olympics_combined['Bronze.1']
olympics_combined['Total_All_Medals'] = olympics_combined['Total_Summer_Medals'] + olympics_combined['Total_Winter_Medals']

# Display the results with relevant columns
olympics_combined[['Country', 'Summer Olympics', 'Total_Summer_Medals', 'Winter Olympics', 'Total_Winter_Medals', 'Total_All_Medals']].head(20)
```
