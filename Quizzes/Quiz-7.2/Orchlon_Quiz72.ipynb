## 2. Training the K-NN model on the Training set

```python
class KNN():
    def __init__(self,k):
        self.k=k
        print('Input k value: ',self.k)
        
    def fit(self,X_train,y_train):
        self.x_train=X_train
        self.y_train=y_train
    
    def calculate_euclidean(self,sample1,sample2):
        distance=0.0
        for i in range(len(sample1)):
            #Euclidean Distance = sqrt(sum i to N (x1_i â€“ x2_i)^2)
            distance+=(sample1[i]-sample2[i])**2 
        return np.sqrt(distance)

    def nearest_neighbors(self, test_sample):
        distances = []  # List to store distances from the test sample to each training sample

        # Loop over each sample in the training set
        for i in range(len(self.x_train)):
            # Calculate the Euclidean distance from the current training sample to the test sample
            # Append a tuple of (training label, distance) to the distances list
            distances.append((self.y_train[i], self.calculate_euclidean(self.x_train[i], test_sample)))

        # Sort the distances list in ascending order based on the distance
        distances.sort(key=lambda x: x[1])

        neighbors = []  # List to store the nearest neighbors

        # Retrieve the first 'k' nearest neighbors
        for i in range(self.k):
            # Append the label of each neighbor to the neighbors list
            neighbors.append(distances[i][0])

        # Return the list of nearest neighbors
        return neighbors

    def predict(self, test_set):
        predictions = []  # List to store predictions for each test sample

        # Loop over each sample in the test set
        for test_sample in test_set:
            # Find the nearest neighbors for the current test sample
            neighbors = self.nearest_neighbors(test_sample)

            # Extract the labels of the nearest neighbors
            labels = [sample for sample in neighbors]

            # Determine the most common label among the neighbors
            prediction = max(labels, key=labels.count)

            # Append the predicted label to the predictions list
            predictions.append(prediction)

        # Return the list of predictions for the test set
        return predictions
```

```python
model=KNN(5) #our model
model.fit(X_train,y_train)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
 
data = {
    'Student': ['A','B','C','D','E','F','G'],
    'Study_Hours': [2,3,4,6,7,8,9],
    'Sleep_Hours': [9,8,7,6,5,4,3],
    'Pass': ['No','No','No','Yes','Yes','Yes','Yes']
}
```

### Steps 1, 2, 3:

```python
data = pd.DataFrame(data)

# Split into features and labels
X = data[['Study_Hours', 'Sleep_Hours']]
y = data['Pass']

# For this small dataset (7 samples), we'll use all data for training
# Since we have only 7 samples, using k=5 means we'd use almost all neighbors
# Let's use k=3 which is more reasonable for small datasets
X_train = X
y_train = y

# Train KNN classifier with k=3 (more appropriate for small dataset)
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)

print(f"Training set: {len(X_train)} samples")
print("Data preview:")
print(data)
```

### 4. Predict for a new student


```python
# Predict for a new student with Study_Hours=5 and Sleep_Hours=6
new_student = [[5, 6]]
prediction = knn.predict(new_student)
probability = knn.predict_proba(new_student)

print(f"New student: Study Hours = 5, Sleep Hours = 6")
print(f"Prediction: {prediction[0]}")
print(f"Probability: Pass = {probability[0][1]:.2f}, Fail = {probability[0][0]:.2f}")
```

### 5. Try different values of K and compare results


```python
# Test different values of k
k_values = [1, 3, 5]
print("Comparing different K values:\n")

for k in k_values:
    knn_temp = KNeighborsClassifier(n_neighbors=k)
    knn_temp.fit(X_train, y_train)
    new_pred = knn_temp.predict(new_student)
    prob_temp = knn_temp.predict_proba(new_student)
    
    print(f"K = {k}:")
    print(f"  New student prediction: {new_pred[0]} (Pass prob: {prob_temp[0][1]:.2f})")
    print()
```

### 6. Visualize the data points with color-coded classes


```python
# Create color mapping
color_map = {'No': 'red', 'Yes': 'green'}

# Plot the data points
plt.figure(figsize=(10, 6))

for label in ['No', 'Yes']:
    mask = data['Pass'] == label
    plt.scatter(data[mask]['Study_Hours'], data[mask]['Sleep_Hours'], c=color_map[label], s=100, alpha=0.7, edgecolor='black', label=f'Pass={label}', linewidth=1.5)

# Plot the new student with prediction
pred_color = 'green' if prediction[0] == 'Yes' else 'red'
marker_style = '^' if prediction[0] == 'Yes' else 'v'
plt.scatter(new_student[0][0], new_student[0][1], 
            c=pred_color, s=200, alpha=0.9, marker=marker_style, 
            edgecolor='black', linewidth=2, label=f'New Student (Predicted: {prediction[0]})')

plt.title('Orchlon - KNN Classification: Study vs Sleep Hours')
plt.xlabel('Study Hours')
plt.ylabel('Sleep Hours')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```
